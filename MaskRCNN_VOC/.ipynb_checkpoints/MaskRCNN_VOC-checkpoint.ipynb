```python
import os
import math
import torch
import torchvision
import numpy as np

import bbox_visualizer as bbv
import matplotlib.pyplot as plt

from PIL import Image
from torchvision.datasets.voc import *
from torchvision.models.detection import FasterRCNN
from torchvision.models.detection.rpn import AnchorGenerator
```

```python
VOC_LABEL_PAIR = {
    'aeroplane': 1,
    'bicycle': 2,
    'bird': 3,
    'boat': 4,
    'bottle': 5, 
    'bus': 6,
    'car': 7,
    'cat': 8,
    'chair': 9,
    'cow': 10,
    'diningtable': 11, 
    'dog': 12,
    'horse': 13,
    'motorbike': 14, 
    'person': 15,
    'pottedplant':16, 
    'sheep': 17,
    'sofa': 18,
    'train': 19,
    'tvmonitor': 20 }


VOCDetection._SPLITS_DIR = "Segmentation"
class VOC_Dataset(torch.utils.data.Dataset):
    def __init__(
        self,
        root: str,
        year: str = "2012",
        image_set: str = "train",
        download: bool = False,
        option = 1
    ):
        transform = torchvision.transforms.Compose([torchvision.transforms.ToTensor()])
        target_transform = torchvision.transforms.Compose([np.array])
        
        VOCSegmentation._TARGET_DIR = "SegmentationClass"
        self.segmentation_dataset = VOCSegmentation(root, year, image_set, download, transform, target_transform, )
        
        VOCSegmentation._TARGET_DIR = "SegmentationObject"
        self.instance_dataset = VOCSegmentation(root, year, image_set, download, transform, target_transform, )
        
        
        self.detection_dataset = VOCDetection(root, year, image_set, download)
    
        self.option = option
    def __len__(self):
        return len(self.segmentation_dataset)
    
    def __getitem__(self, index):
        if self.option == 1:
            image, mask = self.segmentation_dataset[index]    
        else:
            image, mask = self.instance_dataset[index]
#         print(mask.shape)
        
        _, annotation = self.detection_dataset[index]
        
        objects = annotation["annotation"]["object"]
        
        labels = np.array([VOC_LABEL_PAIR[o["name"]] for o in objects], dtype = np.int64)
        boxes = [o["bndbox"] for o in objects]
        boxes = np.array([[box["xmin"], box["ymin"], box["xmax"], box["ymax"]] for box in boxes], dtype = np.float32)
        
        if self.option == 1:
#             print(labels)
            mask = (mask == labels.reshape(-1, 1, 1)).astype(np.uint8) 
        else:
#             print(np.unique(mask)[1:-1])
#             mask = (mask == np.unique(mask)[1:-1].reshape(-1, 1, 1)).astype(np.uint8)
#             print(mask[None, :].shape)
#             print(np.unique(mask)[1:-1])
            id_list = np.unique(mask)
#             print(id_list)
            id_list = np.delete(id_list, np.where((id_list == 0) | (id_list == 255)))
            mask = (mask[None, :] == id_list.reshape(-1, 1, 1)).astype(np.uint8)
            mask = np.ones(mask.shape).astype(np.uint8)
            # shape가 문제다.
    
        target = {
            "masks":torch.tensor(mask),
            "boxes" : torch.tensor(boxes),
            "labels" : torch.tensor(labels)
        }
        return (image, target)

    def get_sample_image_with_bbox(self, sample_idx):
        sample = self[sample_idx]
        image = self.get_image_from_sample(sample)
        boxes = self.get_boxes_from_sample(sample)
        return self.draw_boxes(image, boxes)

    def get_image_from_sample(self, sample):
        return ImageManager(sample[0])
    
    def get_boxes_from_sample(self, sample):
        return BoxManager.tensor_to_list(sample[1]["boxes"])
        
    def draw_boxes(self, image, boxes):
        return bbv.draw_multiple_rectangles(image, boxes)
    
    @staticmethod
    def to(data, device):
        if isinstance(data_sample, list):
             return [to(sample) for sample in data]
        else:
            image, target = data
            imaage = image.to(device)
            target = {
                "masks": target["masks"].to(device),
                "boxes": target["boxes"].to(device),
                "labels": target["labels"].to(device)
            }
            return (image, target)
        
class VOC_Dataloader(torch.utils.data.DataLoader):
    def __init__(self, dataset, batch_size=1, shuffle = False):
        super().__init__(dataset = dataset, batch_size = batch_size, shuffle = shuffle, collate_fn=VOC_Dataloader.collate_batch)
    
    @staticmethod
    def collate_batch(batch):
        images = [sample[0] for sample in batch]
        targets = [sample[1] for sample in batch]
        return images, targets
```

```python
class ImageManager():
    def tensor_to_numpy(image):
        return np.array(image.to("cpu")).transpose((1, 2, 0))
```

```python
class BoxManager():
    @staticmethod
    def tensor_to_list(boxes):
        return boxes.to("cpu").type(torch.uint8).tolist()
    
```

```python

def train_one_epoch(model, optimizer, data_loader, device, epoch, print_freq, scaler=None):
    model.train()
#     metric_logger = utils.MetricLogger(delimiter="  ")
#     metric_logger.add_meter("lr", utils.SmoothedValue(window_size=1, fmt="{value:.6f}"))
#     header = f"Epoch: [{epoch}]"

    lr_scheduler = None
    if epoch == 0:
        warmup_factor = 1.0 / 1000
        warmup_iters = min(1000, len(data_loader) - 1)

        lr_scheduler = torch.optim.lr_scheduler.LinearLR(
            optimizer, start_factor=warmup_factor, total_iters=warmup_iters
        )

#     for images, targets in metric_logger.log_every(data_loader, print_freq, header):
    losses_list = []
    for images, targets in data_loader:
        images = list(image.to(device) for image in images)
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]
        with torch.cuda.amp.autocast(enabled=scaler is not None):
            loss_dict = model(images, targets)
#             losses = sum(loss for loss in loss_dict.values())
            loss_list = torch.stack([value for value in loss_dict.values()])
            losses = loss_list @ loss_list**(1/2)
            losses_list.append(losses)

        # reduce losses over all GPUs for logging purposes
#         loss_dict_reduced = utils.reduce_dict(loss_dict)
        
#         losses_reduced = sum(loss for loss in loss_dict_reduced.values())
#         losses_reduced = sum(loss for loss in loss_dict_reduced.values())
#         loss_value = losses_reduced.item()

#         if not math.isfinite(loss_value):
#             print(f"Loss is {loss_value}, stopping training")
# #             print(loss_dict_reduced)
#             sys.exit(1)

        optimizer.zero_grad()
        if scaler is not None:
            scaler.scale(losses).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            losses.backward()
            optimizer.step()

        if lr_scheduler is not None:
            lr_scheduler.step()
    return torch.mean(torch.stack(losses_list)).to("cpu")



    #     return image
#         metric_logger.update(loss=losses_reduced, **loss_dict_reduced)
#         metric_logger.update(lr=optimizer.param_groups[0]["lr"])

#     return metric_logger
```
