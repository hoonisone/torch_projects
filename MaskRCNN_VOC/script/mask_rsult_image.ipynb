```python
import matplotlib.pyplot as plt
import numpy as np
import bbox_visualizer as bbv

# from ultralytics import YOLO
# from ultralytics.yolo.utils.ops import scale_image
import cv2
import numpy as np
from pathlib import Path

from PIL import Image
from tqdm import tqdm

import mh_utils as MH
from VOC2012_MaskRCNN_InstanceSegmentation import *
from torch.nn.parallel import DistributedDataParallel as DDP
```

```python

def eval_and_save_result(task, save_dir, device):
    path = Path(f"{save_dir}")
    path.mkdir(parents=True, exist_ok=True)
    
    model = task.get_model().eval().to(device)
    _, t_dataloader, _, v_dataloader = task.get_simple_dataset(batch_size = 5)
    idx = 0
    for batch_data in tqdm(v_dataloader):
        batch_data = task.to_for_batch_data(batch_data, device)
        batch_result = task.feed(model, batch_data)
        
        images, targets = Converter.tensor_to_numpy_for_batch_data(batch_data)
        results = Converter.tensor_to_numpy_for_batch_result(batch_result)
        
        for image, result in zip(images, results):
            # 탐지 결과 다듬기
            result = MH.trim(result, score_threshold = 0.7, mask_threshold = 0.5, iou_threshold = 0.3)
            
            masked_image = MH.aggregate(image, result["masks"], alpha = 0.1)
            plt.imshow(masked_image)
            plt.axis('off')
            plt.savefig(path/f"masked_image({idx}).png")


            white_canvas = np.ones(image.shape, dtype = np.float32)
            masked_canvas = MH.aggregate(white_canvas, result["masks"], alpha = 1)
            plt.imshow(masked_canvas)
            plt.axis('off')
            plt.savefig(path/f"masked_canvas({idx}).png")
            
            idx += 1
    
task = MaskRCNN("./../result/0/benchmark/checkpoint(20)")
save_dir = "./../result/0/benchmark/mask_images"
eval_and_save_result(task, save_dir, device = 5)
```
